```python

# RequestsとBeautiful Soupをインポート
import requests
from bs4 import BeautifulSoup

# 記事全文を取得する対象のURL
url ='https://news.yahoo.co.jp/articles/be6fa9e15eb3b41706c7ddf32a75f9b540b9d603'

# GETリクエストを送信し、HTMLを取得する
response = requests.get(url)

# HTMLを解析する
soup = BeautifulSoup(response.text, 'html.parser')

# CSSセレクタを使用して、記事全文を取得する。
article_element = soup.select_one('#uamods > div.article_body.highLightSearchTarget > div:nth-child(1) > p')



# 記事全文のテキストを取得する
article_text = article_element.text

# 取得した記事全文のテキストを出力する
print(article_text)
```
