```python
import requests
import pprint
from getpass import getpass
```

```python
# APIキーを指定
api_key = getpass('APIキーを入力してください')

# 駅名を入力する
station_name = input('検索する駅を入力してください:')

# google map　Geocodingu APIのURL
geocoding_url = 'https://maps.googleapis.com/maps/api/geocode/json'

```

```python
# 駅名から緯度と経度を取得
geocoding_response =requests.get(
    geocoding_url,
    params={
        'address':station_name, # 駅名をパラメータに追加
        'key':api_key,    # APIキーをパラメータに追加
        'language' : 'Japnese'
           }
    )

geocoding_data =geocoding_response.json()

if geocoding_data['status'] == 'OK':
  latitude = geocoding_data['results'][0]['geometry']['location']['lat']
  longitude = geocoding_data['results'][0]['geometry']['location']['lng']

# google mapのAPIのURL
  api_url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'

# データを取得
response = requests.get(
    api_url,
    params={
        'location': f'{latitude},{longitude}',  # 緯度と経度を指定
        'key': api_key,
        'language': 'Japanese',
        'radius': 500,
        'type':'restaurant'
    }
)

#  レスポンスのJSONを取得
json_data = response.json()

# レストラン情報（レストラン名、評価、住所）を抽出
for results in json_data['results']:
  name = results.get('name')
  rating = results.get('rating')
  address = results.get('vicinity')  # 住所は vicinity に格納されている

  if name and rating and address:
    print(f"レストラン名:{name}")
    print(f"評価:{rating}")
    print(f"住所:{address}")
    print("-" * 20)

# JSONの中身を確認
pprint.pprint(json_data)
```
